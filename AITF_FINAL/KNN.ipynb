```python
#載入套件
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#匯入資料（但所有變數名稱皆是隱藏，但需要預測最後TARGET CLASS是哪個）
#index_col=0代表設定第一行為index

df = pd.read_excel('top200_training.xlsx')
df_copy = df.copy()
df.head()
```

```python
#由於KNN分類法是藉由確認最近的觀察值來預測分類，因此變數的比例就會影響很大
#載入標準化比例尺(StandardScaler)套件
from sklearn.preprocessing import StandardScaler
columns_to_exclude = ['簡稱', '證券代碼', '年月','ReturnMean_year_Label']

# 排除指定的欄位
features_to_scale = df.drop(columns=columns_to_exclude)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_to_scale)

# 將標準化後的特徵資料轉換為 DataFrame
df_feat = pd.DataFrame(scaled_features, columns=features_to_scale.columns)
df_feat.head()
```

```python
#選擇特定欄位分析
#from sklearn.preprocessing import StandardScaler

# 指定要選擇的欄位
#selected_columns = ['特定欄位1', '特定欄位2', '特定欄位3']

# 選擇指定的欄位
#features_to_scale = df[selected_columns]

#scaler = StandardScaler()
#scaled_features = scaler.fit_transform(features_to_scale)

# 將標準化後的特徵資料轉換為 DataFrame
#df_feat = pd.DataFrame(scaled_features, columns=selected_columns)
#df_feat.head()
```

```python
#將資料分成訓練組及測試組(自動分割(以index做連結))
from sklearn.model_selection import train_test_split

X = df_feat
y = df['ReturnMean_year_Label']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
#使用KNN演算法
from sklearn.neighbors import KNeighborsClassifier

#從k=1開始測試
clf=KNeighborsClassifier(n_neighbors=29,p=2,weights='distance',algorithm='brute')
clf.fit(X_train,y_train)
pred = clf.predict(X_test)
```

```python
#測試KNN演算法的好壞
from sklearn.metrics import classification_report,confusion_matrix
#將實際類別分為真正例（True Positive）、真負例（True Negative）、偽正例（False Positive）和偽負例（False Negative）
print(confusion_matrix(y_test,pred))
```

```python
#模型的精確度、召回率、F1分數和支持數等指標，用來評估模型對於每個類別的預測性能。
print(classification_report(y_test,pred))
```

```python
#pred 是模型在測試集上的預測結果
pred
```

```python
#返回模型在測試集上的預測準確度。它是模型正確預測的樣本數除以總樣本數的比例。
clf.score(X_test,y_test)
```

```python
#可以用來檢視模型是否存在過擬合的情況。
clf.score(X_train,y_train)
```

```python
# 使用 kneighbors 方法取得最近鄰居的索引和距離
# 這裡假設你想查看測試集中第一個樣本的最近鄰居(問)
test_sample_index = 0
n_neighbors = 5  # 可以自行指定想要查看的最近鄰居的數量

distances, indices = clf.kneighbors(X_test.iloc[[test_sample_index]], n_neighbors=n_neighbors)

# 打印最近鄰居的索引和距離
print(f"最近鄰居的索引: {indices}")
print(f"最近鄰居的距離: {distances}")
```

```python
# 使用 kneighbors 方法取得最近鄰居的索引和距離
# n_neighbors 設定為 5 代表找最近的 5 個鄰居
distances, indices = clf.kneighbors(X_test, n_neighbors=5)

# 合併所有測試樣本的最近鄰居索引
all_indices = np.concatenate(indices)
all_distances = np.concatenate(distances)

# 將索引和距離組合成一個 2D 陣列，方便排序
combined_data = np.column_stack((all_indices, all_distances))

# 按照距離重新排序
sorted_combined_data = combined_data[np.argsort(combined_data[:, 1])]

# 取最近鄰居的前5筆
top5_neighbors_combined_indices = sorted_combined_data[:5, 0].astype(int)
top5_neighbors_combined_distances = sorted_combined_data[:5, 1]
# 根據最近鄰居的索引抓取預測值
#top5_neighbors_predictions = pred[top5_neighbors_combined_indices[top5_neighbors_combined_indices < len(pred)]]


# 打印最近鄰居的前5筆索引和距離
print("最近鄰居的前5筆索引:")
print(top5_neighbors_combined_indices)
print("最近鄰居的前5筆距離:")
print(top5_neighbors_combined_distances)
#print("最近鄰居的前5筆預測值:")
#print(top5_neighbors_predictions)
```

```python
# 假設最近鄰居的索引是 indices[0]
nearest_neighbors_indices = top5_neighbors_combined_indices

# 從原始資料 df 中抓取相應的數值
nearest_neighbors_data = df.iloc[nearest_neighbors_indices]

# 打印最近鄰居的資料
print(nearest_neighbors_data)
```
